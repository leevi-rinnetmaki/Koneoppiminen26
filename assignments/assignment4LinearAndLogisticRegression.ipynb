First we make a graph showing the relationship between the price and all the variables given to us.

```python
from ucimlrepo import fetch_ucirepo
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# fetch dataset
real_estate_valuation = fetch_ucirepo(id=477)

# data (as pandas dataframes)
X = real_estate_valuation.data.features
y = real_estate_valuation.data.targets

for col in X.select_dtypes(include='number').columns:
    plt.figure()
    plt.scatter(X[col], y)
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.title(f'{col} vs price')
    plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

b0 = model.intercept_
b1 = model.coef_[0]
summedError=0
squaredError=0
for counter in range(len(X)):
    error = model.predict(X)[counter][0]-y.to_numpy()[counter][0]
    summedError += error
    squaredError += error**2
summedError = summedError/len(X)
squaredError = squaredError/len(X)

print(summedError, squaredError)
```
